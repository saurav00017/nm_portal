```python
import pandas as pd
import json
```

```python
df = pd.read_csv('./assessment_data.csv')
```

```python
df.head()
```

```python
def check_row(row):
    try:
        data = json.load(row['assessment_data'])
        return data
    except Exception as e:
        return str(e)
```

```python
df['score_check'] = df.apply(check_row, axis=1)
```
