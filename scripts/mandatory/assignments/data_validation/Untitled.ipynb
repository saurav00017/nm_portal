```python
import pandas as pd
import json
import numpy as np
```

```python
df = pd.read_csv('./assessment_data.csv')
```

```python
df = df.dropna(subset=['assessment_data'])
df.head()
```

```python
def check_row(row):
    try:
        data = json.loads(str(row.loc['assessment_data']).replace("'", '"').replace("None", 'null'))
        progress_percentage = row.loc['progress_percentage']
        score_data = {}
        for record in data:
            if 'serial' in record and 'score' in record:
                if record['serial'] in score_data:
                    if score_data[record['serial']] < record['score']:
                        score_data[record['serial']] = record['score']
                    else:
                        score_data[record['serial']] = record['score']
        
        scores_list = score_data.values()
        if scores_list:
            if progress_percentage == sum(scores_list) / len(scores_list):
                return True

            
        return False
    except Exception as e:
        print(e, row)
        return str(e)
```

```python
df.loc['score_check'] = df.apply(check_row, axis=1)
```
